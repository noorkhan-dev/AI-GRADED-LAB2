MIN MAX ALGORATHM

```python
# Minimax implementation for the given tree

def minimax(node, depth, is_maximizing, tree):
    # If node is terminal, return its value
    if node not in tree:
        return node  # terminal value

    if is_maximizing:
        best_val = float('-inf')
        for child in tree[node]:
            val = minimax(child, depth + 1, False, tree)
            best_val = max(best_val, val)
        return best_val
    else:
        best_val = float('inf')
        for child in tree[node]:
            val = minimax(child, depth + 1, True, tree)
            best_val = min(best_val, val)
        return best_val


# Define the tree structure
# Non-terminal nodes map to their children
# Terminal nodes are represented directly by integers
tree = {
    "A": ["B", "C"],         # Maximizer
    "B": [-1, 4, 2],         # Minimizer
    "C": ["F", "G"],         # Minimizer
    "F": [6, -3, -5],        # Maximizer
    "G": [0, 7]              # Maximizer
}

# Run minimax from root "A"
optimal_value = minimax("A", 0, True, tree)
print("Optimal value at root (A):", optimal_value)
```
