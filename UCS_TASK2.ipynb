# ***UCS FOR GRAPH TASK2:***

```python
# Import heapq to use a priority queue (min-heap)
import heapq

def uniform_cost_search(graph, start, goal):
    """
    graph : dictionary representing the graph
    start : starting node
    goal  : target node
    """

    # Priority queue stores tuples of (cost, current_node, path_taken)
    # Start with cost 0, starting node, and empty path
    priority_queue = [(0, start, [])]

    # Visited set to keep track of explored nodes
    visited = set()

    # Loop until the priority queue becomes empty
    while priority_queue:

        # Pop the element with the smallest cost
        cost, current_node, path = heapq.heappop(priority_queue)

        # If node already visited, skip it
        if current_node in visited:
            continue

        # Mark the node as visited
        visited.add(current_node)

        # Add current node to the path
        path = path + [current_node]

        # Print step (for understanding working)
        print(f"Visiting Node: {current_node} | Cost: {cost} | Path: {path}")

        # Goal test: if current node is goal, return result
        if current_node == goal:
            print("\n✅ Goal Reached!")
            return cost, path

        # Explore all neighbors of current node
        for neighbor, weight in graph.get(current_node, []):

            # Only process unvisited neighbors
            if neighbor not in visited:

                # Add neighbor to queue with updated cost
                # New cost = current cost + edge weight
                heapq.heappush(priority_queue, (cost + weight, neighbor, path))

    # If goal is not reachable
    return float("inf"), []


# ---------------- GRAPH DEFINITION ----------------
# Graph is represented as:
# 'Node': [(Neighbor, Cost), ...]

graph = {
    'A': [('B', 1), ('C', 4)],   # A connects to B and C
    'B': [('E', 3), ('D', 1)],   # B connects to E and D
    'E': [('D', 1)],             # E connects to D
    'C': [('F', 5)],             # C connects to F
    'D': [('G', 2)],             # D connects to G (goal path)
    'F': [('G', 2)],             # F connects to G (alternative path)
    'G': []                      # Goal node has no outgoing edges
}


# ---------------- FUNCTION CALL ----------------

# Define start and goal nodes
start_node = 'A'
goal_node = 'G'

# Call UCS algorithm
cost, path = uniform_cost_search(graph, start_node, goal_node)


# ---------------- FINAL OUTPUT ----------------

print("\n📌 Final Result")
print("Optimal Path:", path)   # Shortest path found
print("Total Cost:", cost)     # Minimum cost
```
